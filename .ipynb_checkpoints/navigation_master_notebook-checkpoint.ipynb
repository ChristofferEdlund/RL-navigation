```python
def convert_to_dict(scores, method, avg=False):
    output = []
    
    if avg:
        scores = moving_average(scores, 20)
    
    for i, score in enumerate(scores):
        out = {"Episode":i, "score": score, "method":method}
        output.append(out)
        
    return output

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
# Add all metrics and plot results

metric_path = Path("resources/metrics")
dqn_score = np.load(metric_path.joinpath("dqn_scores.npy"))
double_dqn_score = np.load(metric_path.joinpath("double_dqn_score.npy"))
prio_dqn_score = np.load(metric_path.joinpath("prio_dqn_score.npy"))
mini_rainbow_score = np.load(metric_path.joinpath("mini_rainbow_score.npy"))

dqn_goal = np.load(metric_path.joinpath("dqn_goal.npy"))
ddqn_goal = np.load(metric_path.joinpath("double_dqn_goal.npy"))
prio_dqn_goal = np.load(metric_path.joinpath("PER_dqn_goal.npy"))
minir_dqn_goal = np.load(metric_path.joinpath("mini_rainbow_dqn_goal.npy"))

```

```python
dqn = convert_to_dict(dqn_score, "dqn", True)
ddqn = convert_to_dict(double_dqn_score, "double dqn", True)
pdqn = convert_to_dict(prio_dqn_score, "PER dqn", True)
minir = convert_to_dict(mini_rainbow_score, "mini-rainbow", True)


all_data = dqn + ddqn + pdqn + minir

data = pd.DataFrame(all_data)
```

```python
sns.set_theme()
sns.set(font_scale=1.5, rc={'figure.figsize':(11.7,8.27),"lines.linewidth": 2.5})
sns.set_style("white")


palette = sns.color_palette("BrBG", 4)

ax = sns.lineplot(data=data, x="Episode", y='score', hue="method")
ax.set_title("Navigation agent training plot")

plt.axvline(dqn_goal, 0.0, 1, color="#1f77b4", linestyle='--')
plt.axvline(ddqn_goal, 0.0, 1, color="#ff7f0e", linestyle='--')
plt.axvline(prio_dqn_goal, 0.0, 1, color="#2ca02c", linestyle='--')
plt.axvline(minir_dqn_goal, 0.0, 1, color="#d62728", linestyle='--')
sns.despine(top = True, right = True)

ax.text(dqn_goal-35, 6, (dqn_goal), fontsize=14)
ax.text(ddqn_goal-35, 6, (ddqn_goal), fontsize=14)
ax.text(prio_dqn_goal+5, 6, (prio_dqn_goal), fontsize=14)
ax.text(minir_dqn_goal+-35, 6, (minir_dqn_goal), fontsize=14)
ax.text(300, 4,'Goal reached', fontsize=14)


plt.savefig("resources/plots/smoothed_comparison_plot.png", dpi=300)
```
